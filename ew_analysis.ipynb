```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import igrins_mod as ig # Custom module file for igrins shenanigans

import glob
import os
import warnings

import astropy.units as u


# Plotting Parameters
plt.rcParams['font.size'] = 20
plt.rcParams['axes.labelsize'] = 18
plt.rcParams['xtick.labelsize'] = 18
plt.rcParams['ytick.labelsize'] =18

plt.rcParams['legend.fontsize'] = 16
plt.rcParams['figure.titlesize'] = 20

plt.rcParams['axes.labelweight']='bold'
plt.rcParams['axes.linewidth'] = 3

plt.rcParams['xtick.major.size'] = 10
plt.rcParams['xtick.minor.visible'] = True
plt.rcParams['xtick.minor.size'] = 5

plt.rcParams['ytick.major.size'] = 10
plt.rcParams['ytick.minor.visible'] = True
plt.rcParams['ytick.minor.size'] = 5

plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['xtick.direction'] = 'in'

%matplotlib inline
```

```python
standard_table = pd.read_csv('standard_table_v2.txt', index_col=0)
spec_types = standard_table['Spectral_Type']

hops_table = pd.read_csv('hops_table_v2.txt', index_col=0)
```

```python
# Na I Doublet
st_na_2_2062 = standard_table['ew_na1_2_2062']
st_na_2_2089 = standard_table['ew_na1_2_2089']
st_na_ratio = st_na_2_2062/st_na_2_2089

# Sc and Si
st_sc_2_2058 = standard_table['ew_sc1_2_2058']
st_si_2_2068 = standard_table['ew_si1_2_2068']

# Mg I and Al I
st_mg_2_1065 = standard_table['ew_mg1_2_1065']
st_al_2_1098 = standard_table['ew_al1_2_1098']
st_mg_al_ratio = st_mg_2_1065/st_al_2_1098

# # Ti I and Fe I
# standard_table['ew_']

# Ca I and Fe I
st_ca_2_2614 = standard_table['ew_ca1_2_2614']
st_ca_2_2631 = standard_table['ew_ca1_2_2631']
st_ca_2_2657 = standard_table['ew_ca1_2_2657']
st_fe_2_2626 = standard_table['ew_fe1_2_2626']
```

```python
fig = plt.figure(figsize=(15,5))

plt.plot(spec_types,st_na_2_2062, ls='--', marker='o', label=rf'Na I 2.2062 $\mu$m')
plt.plot(spec_types,st_na_2_2089, ls='--', marker='o', label=rf'Na I 2.2089 $\mu$m')

plt.plot(spec_types,st_sc_2_2058, ls='--', marker='o', label=rf'Sc I 2.2058 $\mu$m')

plt.plot(spec_types,st_si_2_2068, ls='--', marker='o', label=rf'Si I 2.2068 $\mu$m')



plt.xlabel('Spectral Type')
plt.ylabel(rf'EW [$\mu$m]')
plt.title('Na Interval Equivalent Widths')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))


plt.plot(spec_types,st_ca_2_2614, ls='--', marker='o', label=rf'Ca I 2.2614 $\mu$m')
plt.plot(spec_types,st_ca_2_2631, ls='--', marker='o', label=rf'Ca I 2.2631 $\mu$m')
plt.plot(spec_types,st_ca_2_2657, ls='--', marker='o', label=rf'Ca I 2.2657 $\mu$m')
plt.plot(spec_types,st_fe_2_2626, ls='--', marker='o', label=rf'Fe I 2.2626 $\mu$m')

plt.xlabel('Spectral Type')
plt.ylabel(rf'EW [$\mu$m]')
plt.title('Ca Interval Equivalent Widths')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))
plt.plot(spec_types,st_na_ratio, ls='--', marker='o', label=rf'Na I 2.2062/2.2089 $\mu$m')
plt.plot(spec_types,st_mg_al_ratio, ls='--', marker='o', label=rf'Mg I 2.1065/Al I 2.1098 $\mu$m')

# plt.scatter(spec_types,st_sc_2_2058, label=rf'Sc I 2.2058 $\mu$m')
# plt.scatter(spec_types,st_si_2_2068, label=rf'Si I 2.2068 $\mu$m')

plt.xlabel('Spectral Type')
plt.ylabel(rf'EW Ratio')
plt.title('EW Ratios')
plt.legend()
plt.show()
```
