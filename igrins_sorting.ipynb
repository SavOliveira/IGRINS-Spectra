```python
from muler.igrins import IGRINSSpectrum, IGRINSSpectrumList
from specutils import Spectrum1D, SpectrumList

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import glob
import os
from tqdm import tqdm


from astropy.io import fits
```

```python
standard_table = pd.read_csv('./standard_table_v3.txt', index_col=0)  # csv of standards with file and Spectral Type, c/v TBA
standards_path = standard_table['File']
```

```python
standards_path[4]
```

```python
spec = IGRINSSpectrumList.read(standards_path[4])
spec.remove_nans().trim_edges(limits=(210,2000)).plot()
plt.axvline(21075,color='k')

plt.xlim(21100,21650)
plt.show()
```

```python
spec = IGRINSSpectrumList.read(standards_path[4])
spec.trim_edges(limits=(230,2000)).stitch().plot()

plt.axvline(21075,color='k')

plt.xlim(21075-30, 21075+30)
plt.show()
```
